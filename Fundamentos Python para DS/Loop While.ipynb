**<h1> Loop While**

```python
# Usando o loop while para imprimir os valores de 0 a 9
# A condição tem que deixar de ser verdadeira dentro do loop senão ela pode entrar em loop infinito e travar o computador.
valor = 0
while valor <= 10:
  print(valor)
  valor += 1
```

```python
# Também é possível usar a cláusula else para encerrar o loop while

x = 0

while x < 10:
  print('O valor nesta iteração é ',x)
  print('Portanto ainda é menor que 10, somando 1 a x')
  x += 1
else:
  print('Loop concluído')
print(x)
```

**Pass, Break, Continue**

```python
# Se encontrarmos o número 4, interrompemos o loop
valor = 0
while valor < 10:
  if valor == 4:
    print('achou o 4')
    break

  else:
    pass
    print(valor)
    valor += 1
```

```python
# Desconsideramos a letra z ao imprimir os caracteres da frase
for letra in 'Python é zzz incrivel':
  if letra == 'z':
   continue
  print(letra)
```
