**<h1>Fazendo Split dos Dados**

```python
def split_palavras(text):
  return text.split(' ')
```

```python
texto = 'Essa função será bastante útil para separar grandes volumes de dados'
```

```python
# Isso divide a string em uma listra de palavras
print(split_palavras(texto))
```

```python
token = split_palavras(texto)
```

```python
#Fazendo split dos dados
def split_letras(text):
  texto =  text.upper()
  for i in texto:
    print(i,end=' ')
```

```python
split_letras(texto)
```
