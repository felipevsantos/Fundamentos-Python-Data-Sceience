**<h1>Trabalhando com Tuplas**

```python
#criando uma tupla
tupla1 = ('Geografia',23,'Elefantes',9.8)
```

```python
print(tupla1)
```

```python
#tuplas nao suportam append - tuplas são imutáveis
tupla1.append('Chocolate')
```

```python
#tuplas não suportam delete de um item especifico
del tupla1('Elefantes')
```

```python
# Tuplas podem ter um único item
tupla1 = ('Chocolate')
```

```python
print(tupla1)
```

```python
# Usando a função list() para converter uma tupla para lista
lista2 = list(tupla1)
```

```python
print(lista2)
```

```python
lista2.append('s')
```

```python
# Usando a função tuple para converter lista para tupla
tupla1 = tuple(lista2)
```

```python
print(tupla1)
```
