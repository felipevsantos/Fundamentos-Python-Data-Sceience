**<h2> Aplicando Linguagem SQL Direto no Banco de Dados com Linguagem Python</h2>

A query abaixo retorna a média de unidades vendidas

```python
# cria uma instrução SQL para calcular a média de unidades vendidas
query2  = 'SELECT AVG(Unidades_Vendidas) FROM tb_vendas_dsa'
```

```python
# Executa a query no banco de dados
cursor.execute(query2)
```

```python
# Visualiza os Resultados
print(cursor.fetchall())
```

A query abaixo retorna a média de unidades vendidas por produtos

```python
# cria uma query SQL para calcular a media de unidades vendidas por produto
query3 = 'SELECT Nome_Produto,AVG(Unidades_Vendidas) FROM tb_vendas_dsa GROUP BY Nome_Produto'
```

```python
cursor.execute(query3)
```

```python
print(cursor.fetchall())
```

A query abaixo retorna a média de unidades vendidas por produto com valor unitario for maior do que 199

```python
#Cria uma instrução SQL para calcular a média de unidades vendidas por produto,
# quando o valor unitário for maior que 199

query4 = ''' SELECT Nome_Produto, AVG(Unidades_Vendidas)
           FROM tb_vendas_dsa
           WHERE Valor_Unitario > 199
           GROUP BY Nome_Produto'''
```

```python
# Executa a query no banco de dados
cursor.execute(query4)
```

```python
# Visualiza o resultado
print(cursor.fetchall())
```

A query abaixo retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199 e somente ser a média das unidades vendidas for maior do que 10

```python
# Cria instrução SQL para calcular a média de unidades vendidas por produto
#se o valor unitario for maior do que 199 e somente ser a média das unidades vendidas for maior do que 10
```

```python

query5 = ''' SELECT Nome_Produto, AVG(Unidades_Vendidas)
           FROM tb_vendas_dsa
           WHERE Valor_Unitario > 199
           GROUP BY Nome_Produto
           HAVING AVG(Unidades_Vendidas) > 10 '''
```

```python
cursor.execute(query5)
```

```python
print(cursor.fetchall())
```

```python
#Fecha cursor e conexão
cursor.close()
con.close()
```

**<h2> Aplicando Linguagem SQL na Sintaxe do Pandas com Linguagem Python </h2>**

```python
#conecta no banco de dados
con = sqlite3.connect('cap12_dsa.db')
```

```python
#abre um cursor para percorrer os registros do banco de dados
cursor = con.cursor()
```

```python
query = 'SELECT * FROM tb_vendas_dsa'
```

```python
cursor.execute(query)
```

```python
print(cursor.fetchall())
```

```python
cursor.execute(query)
```

```python
dados = cursor.fetchall()
```

```python
# Carrega os dados como dataframe no Pandas

df = pd.DataFrame(dados,columns = ['ID_Pedido',
                                   'ID_Cliente',
                                   'Nome_Produto',
                                   'Valor_Unitario',
                                   'Unidades_Vendidas',
                                   'Custo'])
```

```python
df.head()
```

```python
cursor.close()
con.close()
```

A query abaixo retorna a media de unidades vendidas

```python
media_unidades_vendidas = df['Unidades_Vendidas'].mean()
```

```python
print(media_unidades_vendidas)
```

A query abaixo retorno a media de unidades vendidas por produto

```python
# Calcula a media de unidades vendidas por produto
media_unidades_vendidas_produto = df.groupby('Nome_Produto')['Unidades_Vendidas'].mean()
```

```python
# Visualiza os 10 primeiros resultados
print(media_unidades_vendidas_produto.head(10))
```
