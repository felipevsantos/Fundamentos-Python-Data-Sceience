```python
#input - Ao utilizar essa função, ele pega o que foi digitado e converte para string automaticamente
# precisa entao fazer conversão
idade = int (input("Digite sua idade: "))
if idade > 13:
  print('Você pode acessar redes sociais sem supervisão')
else:
  print('Você deveria acessar redes sociais com supervisão')
```
