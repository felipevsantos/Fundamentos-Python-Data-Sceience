**<h1>Trabalhando com Listas**

```python
# criando uma lista
lista_1 = ['arroz','frango','feijao','tomate']
```

```python
print(lista_1)
```

```python
lista_2 = [12,54,'felipe']
```

```python
print(lista_2)
```

```python
# Atribuindo cada valor da lista a uma variavel

item1 =  lista_2[0]
item2 =  lista_2[1]
item3 =  lista_2[2]
```

```python
print(str(item1) + '-' + str(item2) + '-' + str(item3))
```

```python
# Atualizando um item da lista

lista_2[2] = 'chocolate'
```

```python
print(lista_2)
```

```python
# Deletando um item da lista utiliza palavra reservada del
del lista_2[1]
```

**<h1> Funções Built-in**

```python
#criando uma lista
lista_num =[10,25,41,8.9,3.5]
```

```python
# Função len() retorna o comprimento da lista
print(len(lista_num))
```

```python
# Função max() retorna o valor maximo da lista
print(max(lista_num))
```

```python
# Função min() retorna o valor mínimo da lista
print(min(lista_num))
```

```python
# Criando uma lista
lista_formacoes = ['Engenheiro de Dados', 'Analista de Dados', 'Cientista de Dados']
```
